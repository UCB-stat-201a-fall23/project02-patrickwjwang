### 0. Setup 

We will consider $X_1, X_2, \ldots, X_n$ a sequence of i.i.d. random variables with mean $\mathbb{E}[X_1] = \mu$. We define the sample mean as 
$$
\bar X_n = \frac{1}{n} \sum_{i=1}^n X_i
$$
The objective of this project is to study the behavior of $\bar X_n$ as $n \rightarrow \infty$. 

### 1. Law of Large Numbers

Here we are going to focus in the convergence in probability and almost surely of the sample mean of i.i.d. random variables. Remember that a sequence of random variables $Y_1, Y_2, \ldots$ converges in probability to a random variable $Y$ if for every $\epsilon > 0$ we have 
$$
\lim_{n \rightarrow \infty} \mathbb P \left( | Y_n - Y | > \epsilon \right) = 0
$$
The **weak law of large numbers** states 

> The sample mean $\bar X_n$ converges in probability to $\mu$.

The **strong law of large numbers** states a more strict sense in which this convergence happens

> The sample mean $\bar X_n$ converges almost surely to $\mu$. 

Let's see how these two behave numerically.

#### 1.1. Gaussian case

Consider the simple case where $X_i \sim N(0,1)$. Compute the sample mean for increasing values of $n$ and show how the sample mean converges to $0$. To do this, you can simply show a plot of the sample mean as a function of $n$.

````python
# TO DO CELL
n_min, n_max, inc = 50, 50000, 50
n = np.arange(n_min, n_max+1, inc)

# Compute sample mean for each n
sp_mean_lst = []
for N in n:
    sample_mean = np.mean(np.random.normal(loc=0.0, scale=1.0, size=N))
    sp_mean_lst.append(sample_mean)
sp_mean = np.array(sp_mean_lst)
    
# Plot
plt.plot(n, sp_mean, lw=1, label=f"sample mean")

# Labels and title
plt.title(f"Normal Sample Mean from n={n_min} to {n_max}")
plt.xlabel('n')
plt.ylabel('Mean')
plt.legend()
plt.grid(True)
plt.show()
````

#### 1.3. Compute the probability

Now, instead of evaluating the value of $\bar X_n$, compute the probability $\mathbb P \left( | \bar X_n | > \epsilon \right)$ for different values of $n$ and $\epsilon$ and plot the result as a function of $n$. What do you observe? See the next item for an example of the solution for this problem.

````python
# To do
def cal_prob(epsilon, sim, n):
    # Create prob array to store the prob
    prob_ary = np.zeros_like(n)    
    # For each sim, use the same sample
    for s in range(sim):
        sample_s = np.random.normal(loc=0.0, scale=1.0, size=np.max(n))
        for idx, N in enumerate(n):
            sample_mean = abs(np.mean(sample_s[0:N]))
            if sample_mean > epsilon: prob_ary[idx] += 1

    prob_ary = prob_ary / sim
    return(prob_ary)

# Setting n
n_min, n_max, inc = 10, 21000, 1000
n0 = np.arange(n_min, n_max+1, inc)

# Change epsilon
e1, e2, e3, s = 0.05, 0.02, 0.01, 500
prob_e1 = cal_prob(e1, s, n0)
prob_e2 = cal_prob(e2, s, n0)
prob_e3 = cal_prob(e3, s, n0)

# Plot
plt.plot(n0, prob_e1, lw=1.5, label=f"$\epsilon$={e1}", color='red', marker='o', markersize=2.5)
plt.plot(n0, prob_e2, lw=1.5, label=f"$\epsilon$={e2}", color='blue', marker='o', markersize=2.5)
plt.plot(n0, prob_e3, lw=1.5, label=f"$\epsilon$={e3}", color='green', marker='o', markersize=2.5)

# Labels and title
plt.title("Probabilty for Different n and $\epsilon$")
plt.xlabel('n')
plt.ylabel('P$(|\\bar{X_n}| > \\epsilon)$')
plt.xlim(0, n_max)
plt.ylim(0, 0.5)  # set the y limit to 0.5
plt.legend()
plt.show()
````

#### 1.4. Concentration bounds

Repeat the same plot that in the last item and show how it compares with the upper bound obtain from using Chebyshev's inequality
$$
\mathbb P \left( | \bar X_n | > \epsilon \right) \leq \frac{1}{n \epsilon^2}
$$
and the Chernoff bound
$$
\mathbb P \left( | \bar X_n | > \epsilon \right) \leq 2 \exp \left( - \frac{\epsilon^2 n}{2} \right)
$$
At the end of the day, you should obtain something like this

<img src="solution_1_34.png" alt="Solution" />

````python
# To do
def cal_prob(epsilon, sim, n):
    # Create prob array to store the prob
    prob_ary = np.zeros_like(n)    
    # For each sim, use the same sample
    for s in range(sim):
        sample_s = np.random.normal(loc=0.0, scale=1.0, size=np.max(n))
        for idx, N in enumerate(n):
            sample_mean = abs(np.mean(sample_s[0:N]))
            if sample_mean > epsilon: prob_ary[idx] += 1

    prob_ary = prob_ary / sim
    return(prob_ary)

# Setting n
n_min, n_max, inc = 10, 21000, 1000
n0 = np.arange(n_min, n_max+1, inc)

# Setting lines
e2, s = 0.02, 500
prob_e2 = cal_prob(e2, s, n0)
cheb_bd = 1 / (n0 * (e2 ** 2))
cher_bd = 2 * np.exp(-n0 * (e2 ** 2) / 2)

# Plot
plt.plot(n0, prob_e2, lw=1.5, label=f"$\epsilon$={e2}", color='steelblue', marker='o', markersize=2.75)
plt.plot(n0, cheb_bd, lw=1.5, label=f"Chebyshev's inequality", color='orange', linestyle='--')
plt.plot(n0, cher_bd, lw=1.5, label=f"Chernoff bound", color='green', linestyle='--')

# Labels and title
plt.title("Probabilty Bounds")
plt.xlabel('n')
plt.ylabel('P$(|\\bar{X_n}| > \\epsilon)$')
plt.xlim(0, n_max)
plt.ylim(0, 0.5)  # set the y limit to 0.5
plt.legend()
plt.show()
````

### 2. Central Limit theorem 

Let's first state one more time the Central Limit Theorem

> Consider $X_1, X_2, \ldots, X_n$ a sequence of i.i.d. random variables with mean $\mathbb{E}[X_1] = \mu$ and finite variance $\mathbb{V}ar[X_1] = \sigma^2$. If call 
$$
\bar X_n = \frac{1}{n} \sum_{i=1}^n X_i
$$
the sample mean, then 
$$
\sqrt{n} \frac{\bar X_n - \mu}{\sigma}
$$
converges in distribution to the standard Normal distribution $N(0,1)$.



#### 2.1. Continuous case

Pick a continuous random variable with bounded second moment and show that the central limit theorem holds numerically. Sample the sample mean $\bar X_n$ for various numbers of $n$ and show that $\sqrt{n}(\bar X_n - \mu) / \sigma$ converges to the standard normal distribution.

This is how the solution should look like for the mean of Beta distributions and a total of 10000 simulations. 

<img src="solution_2_1.png" alt="Solution" />

````python
# To do
from scipy.stats import norm
def beta_sim(a, b, sim, n):
    mu = a / (a+b)
    sigma = ((a*b) / ((a+b)**2 * (a+b+1))) ** 0.5    
    sp_lst = []
    for s in range(sim):
        sp_mean = np.mean(np.random.beta(a, b, size=n))
        std_sp_mean = (n ** 0.5) * (sp_mean - mu) / sigma
        sp_lst.append(std_sp_mean)
    sp = np.array(sp_lst)
    return(sp)

# Simulate with various n
a, b, s = 2, 5, 5000
n1, n2 = 100, 10000
beta_sim1 = beta_sim(a, b, s, n1)
beta_sim2 = beta_sim(a, b, s, n2)
x = np.linspace(-3, 3, 90)

# Plot hist
plt.hist(beta_sim1, bins=x, color='steelblue', density=True, alpha=0.5, label=f"n={n1}")
plt.hist(beta_sim2, bins=x, color='red', density=True, alpha=0.25, label=f"n={n2}")

# Plot normal distribution
x = np.arange(-3, 3, 0.01)
pdf = norm.pdf(x, 0, 1)
plt.plot(x, pdf, color='orange', lw=2, label=f"N(0, 1)")

# Labels and title
plt.title('Central Limit Theorem (Beta)')
plt.xlabel('$\\sqrt{n}(\\bar{X_n} - \\mu) / \\sigma$')
plt.ylabel('Density')
plt.legend()
plt.show()
````

#### 2.2. Discrete case

Repeat the experiment but now with a discrete random variable. When plotting the histogram, be careful on how you define the bins. 


````python
# To do
from scipy.stats import norm
def bin_sim(n, p, sim, N):
    mu = n * p
    sigma = (n * p * (1-p)) ** 0.5    
    sp_lst = []
    for s in range(sim):
        sp_mean = np.mean(np.random.binomial(n, p, size=N))
        std_sp_mean = (N ** 0.5) * (sp_mean - mu) / sigma
        sp_lst.append(std_sp_mean)
    sp = np.array(sp_lst)
    return(sp)

# Simulate with various n
n, p, s = 10, 0.33, 5000
N1, N2 = 100, 10000
bin_sim1 = bin_sim(n, p, s, N1)
bin_sim2 = bin_sim(n, p, s, N2)
x = np.linspace(-3, 3, 90)

# Plot hist
plt.hist(bin_sim1, bins=x, color='steelblue', density=True, alpha=0.5, label=f"n={n1}")
plt.hist(bin_sim2, bins=x, color='red', density=True, alpha=0.25, label=f"n={n2}")

# Plot normal distribution
x = np.arange(-3, 3, 0.01)
pdf = norm.pdf(x, 0, 1)
plt.plot(x, pdf, color='orange', lw=2, label=f"N(0, 1)")

# Labels and title
plt.title('Central Limit Theorem (Binomial)')
plt.xlabel('$\\sqrt{n}(\\bar{X_n} - \\mu) / \\sigma$')
plt.ylabel('Density')
plt.legend()
plt.show()
````
